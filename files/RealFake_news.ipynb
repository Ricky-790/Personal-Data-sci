```python
import re
true_news['text'] = true_news['text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
fake_news['text'] = fake_news['text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
true_news['text'] = true_news['text'].str.lower()
fake_news['text'] = fake_news['text'].str.lower()

true_news['title'] = true_news['title'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
fake_news['title'] = fake_news['title'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
true_news['title'] = true_news['title'].str.lower()
fake_news['title'] = fake_news['title'].str.lower()
```

##Create a list of all words and extract all the unique words. This will be used for vectorization

```python
list_of_words = []
for i in data['combined']:
  for j in i:
    list_of_words.append(j)
print(len(list_of_words))
list_of_words = set(list_of_words)
```

```python
num_of_words = len(list_of_words)
num_of_words
```

```python
array = []
for i in data['combined']:
  array.append(len(i))

max(array) # This gives the length of the longest sequence
```

##Split into train and test sets

```python
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) #if len(data)=10, shuffled indices will be an array like:[7,4,1,0,8,9..] It will have all numbers from 0 to 10 in a random sequence
    test_set_size=int(len(data)*test_ratio) #Creating the test set
    test_indices=shuffled_indices[:test_set_size]
    train_indices=shuffled_indices[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(data,0.2)
len(train_set), len(test_set)
```

```python
x_train = train_set['combined']
y_train = train_set['T/F']

x_test = test_set['combined']
y_test = test_set['T/F']
```

###Text to sequence

```python
from nltk import word_tokenize
from tensorflow.keras.preprocessing.text import Tokenizer
# Create a tokenizer to tokenize the words and create sequences of tokenized words
tokenizer = Tokenizer(num_words = num_of_words)
tokenizer.fit_on_texts(x_train)
train_sequences = tokenizer.texts_to_sequences(x_train)
test_sequences = tokenizer.texts_to_sequences(x_test)
```

```python
print(f'''
  text = {x_train.iloc[0]}
  converted_sequence ={train_sequences[0]}
''')
```

##Padding the data

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_train = pad_sequences(train_sequences,maxlen = 40, padding = 'post', truncating = 'post')
padded_test = pad_sequences(test_sequences,maxlen = 40, truncating = 'post')
```

##Building the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding, Input, LSTM
from tensorflow.keras.models import Model
```

```python
model = Sequential()

# embedding layer
model.add(Embedding(num_of_words, output_dim = 128))

# Bi-Directional RNN and LSTM
model.add((LSTM(128)))

# Dense layers
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1,activation= 'sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
y_train = np.asarray(y_train)
```

#Training the model

```python
model.fit(padded_train, y_train, batch_size = 64, validation_split = 0.1, epochs = 6)
```

```python
# make prediction
pred = model.predict(padded_test)
```

```python
pred[1]
```

```python
prediction = []
for i in range(len(pred)):
    if pred[i].item() > 0.5:
        prediction.append(1)
    else:
        prediction.append(0)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(list(y_test), prediction)

print("Model Accuracy : ", accuracy)
```
